Handle the Title Column

```python
df['product_title'].unique()
```

```python
def clean_text(text):
    text = str(text)
    text = re.sub(r'[\|\-\,\(\)]', ' ', text)
    text = re.sub(r'\s+', ' ', text)       
    return text.strip()and text.lower()

df["clean_title"] = df["product_title"].apply(clean_text)
```

## Feature Engineering

Brand Freature

```python
brands = ["Samsung", "Apple", "Motorola", "Nokia", "OnePlus", "VTech", "Panasonic",
           "BLU", "Alcatel", "Moto", "Nokia", "Nord", "Galaxy", "Pixel","iPhone",
             "Redmi", "Realme", "Oppo", "Vivo", "Huawei", "Xiaomi","TCL"]

def extract_brand(text):
    text = text.lower()
    words = text.split()
    first_words = words[:3]

    for brand in brands:
        brand_lower = brand.lower()

        if brand_lower in first_words:
            return brand

    return "Not Famous brand"

df["Brand"] = df["clean_title"].apply(extract_brand)
```

Product Type

```python
def extract_type(text):
    text = text.lower()

    if "charger" in text or "power bank" in text:
        return "Charger"
    
    if "cordless phone" in text or "landline phone"in text or "landline phone" in text or "cordless "in text:
        return "Home Phone"
    
    if "holder" in text or "mount" in text:
        return "Holder"
    
    if "tripod" in text:
        return "Tripod"
    
    if "compatible with iphone" in text or "for iphone" in text or "Compatible with Apple iPhone, iPad, and iPod" in text:
        pass 
    elif "iphone" in text:
        return "iPhone"
    
    if "with case" in text or "case bundle" in text or "charger bundle" in text  :
        return "Smartphone with bundle"
    
    if( "case" in text or "cover" in text or 
       "casing" in text or "shell" in text or
        "bumper" in text or "sleeve" in text or "pouch" in text or 
     "holster" in text or"spidercase" in text) : 
        return "Case" 

    
    if ("headphones" in text or "earbuds" in text or 
        "earphones" in text or "headset" in text 
        or "corded speakerphone" in text) :
        return "Headphones"
    
    if "flip phone"in text or "flip"in text:
        return "Flip Phone"
    
    if ("smartphone" in text or "cell phone" in text or "android" in text 
        or "galaxy" in text or "moto" in text or "nord" in text or "nokia" in text 
        or "oneplus" in text or "pixel" in text or "redmi" in text or "realme" in text
        or "oppo" in text or "vivo" in text or "huawei" in text or "xiaomi" in text
        or "samsung" in text or "Flip Phone" in text 
        or " dual sim" in text ):
        return "Smartphone"
    
    
    
    else:
        return "Other"

df["Product_Type"] = df["clean_title"].apply(extract_type)
```

remove outliers

```python
## Exprloring the data cleaning and preprocessing steps
q1 = df['product_price'].quantile(0.25)
q3 = df['product_price'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
outliers = df[(df['product_price'] < lower_bound) | (df['product_price'] > upper_bound)]
```
